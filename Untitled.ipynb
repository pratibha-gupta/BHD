```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=.33,random_state=5)
lm = LinearRegression()
model = lm.fit(x_train,y_train)
```

```python
pred_y = lm.predict(x_test)
```

```python
pd.DataFrame({"Actual": y_test, "Predict": pred_y}).head()
```

```python
from sklearn.metrics import fbeta_score , make_scorer
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_train_predict = lm.predict(x_train)
rmse = np.mean(mean_squared_error(y_train,y_train_predict))
r2 = r2_score(y_train,y_train_predict)
```

```python
print("for training set")
print("rmse: ",rmse)
print("r2: ",r2)
```

```python
y_test_predict = lm.predict(x_test)
rmse = np.mean(mean_squared_error(y_test,y_test_predict))
r2 = r2_score(y_test,y_test_predict)
```

```python
print("for test set")
print("rmse: ",rmse)
print("r2: ",r2)
```
